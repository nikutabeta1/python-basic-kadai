```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
        print(f"{self.name}さんは、{self.age}歳なので大人です。")
    else:
        print(f"{self.name}さんは、{self.age}歳なので大人ではありません。")

human_data = [{"name": "侍太郎", "age": 23},
              {"name": "侍次郎", "age": 16},
              {"name": "侍三郎", "age": 12}]

human_list =[]

for data in human_data:
  human = Human(data["name"], data["age"])
  human_list.append(human)

for human in human_list:
  human.check_adult()
```
